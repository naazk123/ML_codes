##### Implement Gradient Descent Algorithm to find the local minima of a function. For example, find the local minima of the function y=(x+3)Â² starting from the point x=2.

$x_{n + 1} = x_n - \alpha \nabla f(x_n)$

```python
def output(x):
    return ((x + 3) ** 2)

def derivative(x):
    return (2*x + 6)

def main():
    steps = 300
    alpha = 0.1
    x = 2
    
    while(steps>0):
        previous_output = output(x)
        x = x - alpha * derivative(x)
        if output(x) > previous_output:
            x = x + alpha * derivative(x)
            break
        print('x = {:.10f}, y = {:.10f}'.format(x, output(x)))
        steps -= 1
    print('Local minima of function y = {:.10f} at x = {:.10f}'.format(x, output(x)))

main()
```
